```python
def fibo(x):
    if x ==1 or x==2:
        return 1
    else:
        return fibo(x-1)+fibo(x-2)
fibo(10)
```

```python
# 하향식(TopDown 방식)
# 메모이제이션을 사용하여 다이나믹 프로그래밍을 구현하는 방식 중 하나
# 메모이제이션이란 한번 구한 계산 결과를 메모리 공간에 메모해두고, 같은 식을 다시 호출하면
# 메모한 겨로가 그대로 가져오는 기법을 말한다.
```

```python
memo = [0]*100
def fibo(x):
    if x ==1 or x==2:
        return 1
    elif memo[x] != 0:
        return memo[x]
    memo[x] = fibo(x-1)+fibo(x-2)
    return memo[x]
fibo(10)
```

```python
def factorial(n):
    if n == 1:
        return 1
    elif n >1:
        return n*factorial(n-1)
factorial(4)
```
