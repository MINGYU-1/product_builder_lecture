```python
N,M = map(int,input().split())
list_1 = [0]*1000
for _ in range(M):
    sum = 0
    i,j = map(int,input().split())
    for n in list_1[i-1:j]:
        sum += n
    print(sum)

```

```python
def add(a,b):
    print(a+b)
try:
    a,b = map(int,input().split())
    while a !=0 & b != 0:
        add(a,b)
        try:
            a,b = map(int,input().split())
        except:
            break
except:
    pass
```

```python
import sys
a = int(sys.stdin.readline())
data = []
for i in range(a):
    line = int(sys.stdin.readline())
    data.append(line)
data.sort()
for i in data:
    print(i)
```
