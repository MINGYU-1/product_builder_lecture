```python
N = list(map(int,input().rstrip()))
cnt = 0
M = [0]*2 
def cycle(N):
    global cnt
    if len(N) == 1:
        cnt+=1
        return cnt
    elif len(N) == 2:      
        M[1] = (N[0]+N[1]) % 10
        M[0] = N[1]
        cnt+=1
        cycle(M)
cycle(N)
```
