```python
n = int(input())
cnt = 0
for i in range(1,n+1):
    c = 0
    for j in range(i,n+1):
        c += j
        if c== n:
            cnt +=1
            break
        elif c>=n:
            break
```

```python
import sys
input = sys.stdin.readline
N,S = map(int,input().split())
arr = list(map(int,input().split()))
start,end = 0,1
list_1=[]
while end <= N:
    arr_sum = sum(arr[start:end])
    if arr_sum < S:
        end +=1
    elif arr_sum ==S:
        list_1.append(end-start+1)
    elif arr_sum > S:
        start +=1
list_1.sort()
print(list_1[0])
```
